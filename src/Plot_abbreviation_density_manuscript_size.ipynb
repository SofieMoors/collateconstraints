```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
# Create dataframe

manuscript_names = ['A', 'B', 'BR', 'C', 'E', 'F', 'G', 'Ge', 'K', 'L', 'O', 'W', 'Y', 'Z']
manuscript_size_cm2 = [516.75, 408, 321.18, 661.5, 367.2, 494, 157.5, 260, 304.5, 346.32, 272, 494, 343.44, 348]
abbreviation_ratio = [0.954410, 0.959768, 0.954891, 0.985194, 0.906042, 0.979722, 0.888841, 0.961567, 0.954339, 0.961337, 0.981558, 0.931787, 0.973158, 0.905969]

sns.set_style('whitegrid')
data = {'Manuscript Names': manuscript_names, 'Abbreviation Ratio': abbreviation_ratio, 'Manuscript Size (cm²)': manuscript_size_cm2}
df = pd.DataFrame(data)

# Sort the dataframe by manuscript size in ascending order
df = df.sort_values(by='Manuscript Size (cm²)')
#print(df)

# Create regression plot
sns.regplot(x='Manuscript Size (cm²)', y='Abbreviation Ratio', data=df, color= 'orange')

# Set labels and title
plt.xlabel('Manuscript size in cm2')
plt.ylabel('Mean Abbreviation Ratio')
plt.title('Regression Plot')

#plt.xticks(range(len(df['Manuscript Names'])), df['Manuscript Names'])

# Annotate each point with manuscript names
for i in range(len(df)):
    plt.annotate(df['Manuscript Names'][i], (df['Manuscript Size (cm²)'][i], df['Abbreviation Ratio'][i]), xytext=(5, 5), textcoords='offset points')

#plt.show()
plt.savefig('regressionplot.png')
```
